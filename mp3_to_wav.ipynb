```
import subprocess
import glob
```

```
def mp3_to_wav_convertor(name, file_path):
  subprocess.call(
      [
          "ffmpeg",
          "-i",
          file_path + "/" + name + ".mp3",
          file_path + "/" + name + ".wav",
      ]
  )
```

```
file_path = drive_location + "/*.mp3"
list_file_names = glob.glob(file_path)
for file in list_file_names:
  begin_idx = 39
  end_idx = file.find(".")
  file_name = file[begin_idx:end_idx]
  mp3_to_wav_convertor(file_name, drive_location)
```
